```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

Fazendo as funcoes diretamente, para poder entendê-las a fundo.

# Função Step (função degrau)

```python
image = mpimg.imread("../img/step.png")
plt.imshow(image)
plt.show()
```

```python
def stepFunction(valor):
    if(valor >= 1):
        return 1
    return 0
```

```python
teste = stepFunction(20)
teste
```

# Função Sigmoid

```python
image = mpimg.imread("../img/sigmoid.png")
plt.imshow(image)
plt.show()
```

```python
def sigmoidFunction(valor):
    return 1 / (1 + np.exp(-valor))
```

```python
teste = sigmoidFunction(20)
teste
```

# Função Tangent

```python
image = mpimg.imread("../img/tahn.png")
plt.imshow(image)
plt.show()
```

```python
def tahnFunction(valor):
    return (np.exp(valor) - np.exp(-valor)) / (np.exp(valor) + np.exp(-valor))
```

```python
teste = tahnFunction(0.358)
teste
```

# Função Relu

```python
image = mpimg.imread("../img/relu.png")
plt.imshow(image)
plt.show()
```

```python
def reluFunction(valor):
    if(valor >= 0):
        return valor
    return 0
```

# Função Softmax

```python
image = mpimg.imread("../img/softmax.png")
plt.imshow(image)
plt.show()
```

```python
def softmaxFunction(vetor):
    ex = np.exp(vetor)
    return ex / ex.sum()
```

```python
vetor = [5, 2, 1.3]
teste = softmaxFunction(vetor)
teste
```
