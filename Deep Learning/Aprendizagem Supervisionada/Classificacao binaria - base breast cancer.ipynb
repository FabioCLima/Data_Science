```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score

import keras
from keras.models import Sequential
from keras.layers import Dense

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
entradas = pd.read_csv("../data/entradas_breast.csv")
saidas = pd.read_csv("../data/saidas_breast.csv")
```

```python
entradas.head()
```

```python
entradas_train, entradas_test, saidas_train, saidas_test = train_test_split(entradas, saidas, test_size=0.25)
```

```python
image = mpimg.imread("../img/rede neural densa.png")
plt.imshow(image)
plt.show()
```

# Rede Neural com uma camada oculta

```python
classificador = Sequential()
classificador.add(Dense(units=16, activation="relu", kernel_initializer="random_uniform", input_dim = 30))
```

```python
classificador.add(Dense(units = 1, activation="sigmoid"))
```

```python
image = mpimg.imread("../img/rede neural exemplo.png")
plt.imshow(image)
plt.show()
```

```python
classificador.compile(optimizer="adam", loss="binary_crossentropy", metrics=["binary_accuracy"])
```

```python
classificador.fit(entradas_train, saidas_train, batch_size=10, epochs=100)
```

```python
previsoes = classificador.predict(entradas_test)
previsoes = (previsoes > 0.5)
```

```python
previsao = accuracy_score(saidas_test, previsoes)
previsao
```

```python
matriz_confusao = confusion_matrix(saidas_test, previsoes)
matriz_confusao
```

```python
resultado = classificador.evaluate(entradas_test, saidas_test)
resultado
```

# Rede Neural com duas camadas ocultas

```python
classificador = Sequential()
classificador.add(Dense(units=16, activation="relu", kernel_initializer="random_uniform", input_dim = 30))
```

```python
classificador.add(Dense(units=16, activation="relu", kernel_initializer="random_uniform"))
```

```python
classificador.add(Dense(units = 1, activation="sigmoid"))
```

```python
image = mpimg.imread("../img/rede neural densa dupla.png")
plt.imshow(image)
plt.show()
```

```python
optimizador = keras.optimizers.Adam(lr= 0.001, decay= 0.0001, clipvalue= 0.5)
```

```python
classificador.compile(optimizer=optimizador, loss="binary_crossentropy", metrics=["binary_accuracy"])
```

```python
classificador.fit(entradas_train, saidas_train, batch_size=10, epochs=100)
```
