```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import keras
from keras.models import Sequential
from keras.layers import Dense

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
entradas = pd.read_csv("../data/entradas_breast.csv")
saidas = pd.read_csv("../data/saidas_breast.csv")
```

```python
entradas.head()
```

```python
entradas_train, entradas_test, saidas_train, saidas_test = train_test_split(entradas, saidas, test_size=0.25)
```

```python
image = mpimg.imread("../img/rede neural densa.png")
plt.imshow(image)
plt.show()
```

```python
classificador = Sequential()
classificador.add(Dense(units=16, activation="relu", kernel_initializer="random_uniform", input_dim = 30))
```

```python
classificador.add(Dense(units = 1, activation="sigmoid"))
```

```python
image = mpimg.imread("../img/rede neural exemplo.png")
plt.imshow(image)
plt.show()
```

```python
classificador.compile(optimizer="adam", loss="binary_crossentropy", metrics=["binary_accuracy"])
```

```python
classificador.fit(entradas_train, saidas_train, batch_size=10, epochs=100)
```
